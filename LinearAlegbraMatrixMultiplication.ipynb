## Question  

Write the Python code to multiply two matrices without using NumPy libraries. Verify result using NumPy function.

Input:
Number of rows, columns and elements of matrix 1 and Matrix 2.

Output:
The product of two matrices.


## Solution

#### Approach :

1. Write code to take inputs from the user to define the matrices.

2. After taking the inputs, create two empty lists to store the values in it and one null matrix with same dimention to resultant matrix which would be obtained

3. Take inputs from the user about the elements in each matrices

4. Create three loops :

    * *first one to loop through the rows of the first matrix*
    
    * *second one to loop through the columns of the first matrix or row of second matrix*
    
    * *third one to loop through the columns of the second matrix*
    
    
5. Using these loops loop through the elements and mulitply and add them and eventually store the value in null matrix created 


    

***
#### Error Handling:

There are few error handling done by the code:

1. Accidental non numeric input value by user

2. To only take non-zero poisive interger value for defining dimensions of matrices

3. To warn user if the dimension of the matrices are incompatible for getting the product
***


```python
#Take input from user and make sure it is a non-zero positive integer and not punctuations or string

def inputNumber(message):
    while True:
        try:
            userInput = int(input(message))
            if userInput <= 0:
                print("Enter Positive Integer Value!")
                continue
        except ValueError:
            print("Not an integer! Try again.")
            continue
        else:
            return userInput 
            break 

#Take input from user and make sure it is a number and not punctuations or string etc for the input of matrix elements

def inputElement(msg,no_of_elements):     
    while True:
        try:
            userInput = input(msg).split()
            if len(userInput)!= no_of_elements:
                print("No of elements not matching!")
                continue
            else:
                for ele in userInput:
                    float(ele)
        except ValueError:
            print("Not a number! Try again.")
            continue     
        else:
            return list(map(float,userInput)) 
            break    
            

            
# Creating the null Matrix for defined dimension

def createEmptyMatrix(rows,cols):
    mat = []
    for i in range(rows):
        mat.append([])
        for j in range(cols):
            mat[i].append(j*0)
    
    return mat
```

```python
#MAIN PROGRAM STARTS HERE:

mat_a_row = inputNumber("Number of rows in first matrix\n")
mat_a_column = inputNumber("Number of columns in first matrix\n")

mat_b_row = inputNumber("Number of rows in second matrix\n")
mat_b_column = inputNumber("Number of columns in second matrix\n")

if mat_a_column != mat_b_row :
    print('Cross Product not possible due to incompatible matrix dimension')
else:
    # now once the matrix dimension is compatible we need to obtain matrix elements
    # create empty matrix for storing results and empty list for storing elements for input matrices
    
    mxA = []
    mxB = []
    mxC = createEmptyMatrix(mat_a_row,mat_b_column)
    
    print("Enter Elements giving space between each elements")
    
    # Defining elements for two matrices
    
    for i in range(mat_a_row):
        mx = inputElement("Enter elments for first Matrix Row " + str(i+1) + " containing " + str(mat_a_column) + " elements ", mat_a_column)
        mxA.append(mx)
                
    for i in range(mat_b_row):
        mx = inputElement("Enter elments for second Matrix Row " + str(i+1) + " containing " + str(mat_b_column) + " elements ", mat_b_column )
        mxB.append(mx)
    
    print('\n')
    
    # Performing the cross multiplication
    
    for i in range(mat_a_row):
        for j in range(mat_a_column):
            for k in range(mat_b_column):
                mxC[i][k] += mxA[i][j]*mxB[j][k]

print('\n')
    
print("First Matrix : ")
print(mxA) 
print('\n')

print("Second Matrix : ")
print(mxB)
print('\n')

print("Result Matrix : ")
print(mxC)
print('\n')

```

#####   The Solution obtained can be cross checked using the numpy function :

```python
import numpy as np
a = np.array(mxA)
b = np.array(mxB)
c = np.matmul(a,b)

print("First Matrix using numpy : ")
print(a) 
print('\n')

print("Second Matrix using numpy : ")
print(b)
print('\n')

print("Result Matrix using numpy : ")
print(c)
print('\n')
```
